```python


merged_df = pd.read_csv('merged_all_2023_.csv')
data_df = pd.read_csv('prev_onlyIITs_data_new.csv')

final_df = pd.concat([data_df, merged_df], ignore_index=True)
final_df = final_df.drop(columns=['Quota', 'Type'])
final_df = final_df.loc[:, ~final_df.columns.str.contains('^Unnamed')]

final_csv_file_path = 'final_merged_data_only_IIts.csv'
final_df.to_csv(final_csv_file_path, index=False)

print(f"Final merged data has been saved to {final_csv_file_path}")
```
